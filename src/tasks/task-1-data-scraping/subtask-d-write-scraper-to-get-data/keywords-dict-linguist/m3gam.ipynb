```python
alphapet = {
    0 : 'a', 1 : 'b', 2 : 'c', 3 : 'd', 4 : 'e', 5 : 'f', 6 : 'g', 7 : 'h', 8 : 'i', 9 : 'j', 10 : 'k', 11 : 'l', 12 : 'm', 13 : 'n', 14 : 'o', 15 : 'p', 16 : 'q', 17 : 'r', 18 : 's', 19 : 't', 20 : 'u', 21 : 'v', 22 : 'w', 23 : 'x', 24 : 'y', 25 : 'z'
}
```

```python
r = requests.get('https://dict.linguist.page/')
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
table = soup.find_all('tbody')
```

```python
words = {}
for key, i in enumerate(table):
    list = i.find_all('td')
    for index in range(0, len(list), 2):
        # print(list[index].text)
        try: 
            if list[index].text[0].lower() == alphapet[key]:
                words[list[index].text] = list[(index+1)].text
            else :
                print(list[index].text)    
        except:
            print(list[index].text)
```
