```python
import csv
import pandas as pd
from selenium import webdriver
#from webdriver_manager.microsoft import EdgeChromiumDriverManager
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoAlertPresentException
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
from selenium.webdriver.support.select import Select
import time
import requests
```

```python
all_urls= []
chrome_options = Options()
#url = 'https://www.coursera.org/specializations/deep-learning?'
#print(url)
#r = requests.get(url)
#soup = BeautifulSoup(r.content, 'html.parser')
#courses = soup.find_all('a', {'data-e2e': 'course-link'})
#print(len(courses))
courses = ['/learn/neural-networks-deep-learning?specialization=deep-learning', '/learn/deep-neural-network?specialization=deep-learning',
           '/learn/machine-learning-projects?specialization=deep-learning', '/learn/convolutional-neural-networks?specialization=deep-learning',
           '/learn/nlp-sequence-models?specialization=deep-learning']

for course in courses:
    course_url = 'https://www.coursera.org' + course
    print(course_url)
    r = requests.get(course_url)
    soup = BeautifulSoup(r.content, 'html.parser')
    weeks = soup.find_all('a', {'class': 'text-primary link-no-style'})
    for week in weeks:
        week_url = 'https://www.coursera.org' + week.get("href")
        print(week_url)
        r = requests.get(week_url)
        soup = BeautifulSoup(r.content, 'html.parser')
        videos = soup.find('div', {'class': 'other-videos css-17ph8gb'}).find_all('a', {'class': 'cds-1'})
        print(len(videos))
        for video in videos:
            video_url = 'https://www.coursera.org' + video.get('href')
            all_urls.append(video_url)
#print(urls_str) 
```

```python
#driver = webdriver.Edge(EdgeChromiumDriverManager().install())
chrome_options = Options()
driver = webdriver.Chrome(ChromeDriverManager().install(), options=chrome_options)
#driver.get('https://www.coursera.org/?authMode=login')
#page_source = driver.page_source
#print(page_source)
#inputs = driver.find_elements_by_css_selector("input.css-7dsrhu")
#inputs[0].send_keys('user name')
#inputs[1].send_keys('password')
#login_btn = driver.find_elements_by_css_selector("button._6dgzsvq.css-1af0gyj")[0]
#login_btn.click()
#driver.implicitly_wait(30)
#driver.get('https://www.coursera.org/lecture/nlp-sequence-models-ar/lm-lnmdhj-lmt-qb-0h7gT?redirectTo=%2Flearn%2Fnlp-sequence-models-ar%3Faction%3Denroll')
#page_source = driver.page_source
#print(page_source)

#urls = driver.find_elements_by_css_selector("a.cds-1.cds-63.cds-65.other-video-link-cds.css-1i4hn45.cds-24")
#print(len(urls))
#urls_str = []
transcripts = []
#for url in urls:
#    url = url.get_attribute("href")
#    urls_str.append(url)
for url_str in all_urls:
    transcript = {} 
    print(url_str)
    transcript['url'] = url_str
    driver.get(url_str)
    driver.find_elements_by_css_selector('button.tab-button')[1].click()
    try:
        sel = Select(driver.find_element_by_css_selector('select.css-qs4aom'))
        sel.select_by_value('ar')
        time.sleep(2)
        content = driver.find_element_by_css_selector('main.css-12a5gj5')
        #print(content.text)
        transcript['Arabic_transcript'] = content.text
    except:
        transcript['Arabic_transcript'] = ''
    try:    
        sel.select_by_value('en')
        time.sleep(2)
        content = driver.find_element_by_css_selector('main.css-12a5gj5')
        #print(content.text)
        transcript['English_transcript'] = content.text
    except:
        transcript['English_transcript'] = ''
    transcripts.append(transcript)
```
