```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
all_data = []
for page in range(1,34):
    url = 'https://datascience.codata.org/articles/?app=25&%20order=date_published&%20f=8&%20f=2&%20f=3&%20f=4&%20f=5&%20f=6&%20f=7&%20f=9&%20f=10&page='+str(page)
    print(url)
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html.parser')
    articles = soup.find_all('div', {'class': 'caption-text'})
    for article in articles:
        try:
            article_title = article.find('h5').text.strip()
            print(article_title)
            article_url = 'https://datascience.codata.org' + article.find('a').get('href')
            print(article_url)
            r = requests.get(article_url)
            soup = BeautifulSoup(r.content, 'html.parser')
            article = soup.find_all('div',{'class':'article-body'})[-1]
            sections = article.find_all('div')
            text = ''
            for section in sections:
                text += section.text.strip()

            all_data.append({'ArticleTitle':article_title, 'Content':text,'URL':article_url})

            print('******************************************************') 
        except:
            pass
```

```python
df = pd.DataFrame(all_data)
df
```
